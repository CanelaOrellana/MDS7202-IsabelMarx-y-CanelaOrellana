### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [x] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [x] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [x] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [x] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [x] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

````python
#Se calcula el promedio de notas y se guarda en una variable GPA
df_grades['GPA'] = (df_grades['math score']+df_grades['reading score']+df_grades['writing score'])/3 
````

````python
#Diccionario para simplificar los niveles de educación
dic_level_education = {'some high school': 'school',
                       'some college' : 'school',
                       'high school' : 'school',
                       'bachelor\'s degree' : 'college',
                       'associate\'s degree' : 'college',
                       'master\'s degree' : 'postgraduate'}
#Se crea la columna que contiene un nivel de educación simplificado
df_grades['simple parental level of education']=df_grades['parental level of education'].map(dic_level_education)
````

````python
#Se crea el df que represente lo que será la tabla resumen, hasta ahora sólo tiene los promedios
tabla_resumen = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).mean().reset_index()
````

````python
#Se agrega la fila count a la tabla resumen
count = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])['GPA'].count().reset_index(name='count')['count']
tabla_resumen['count'] = count

#Se agrega la fila percentage que contiene el porcentaje de alumnos respecto al total
total = tabla_resumen['count'].sum()
tabla_resumen['percentage'] = round(tabla_resumen['count']*100/total,2).astype(str) + '%'

#Se vuelven a instalar los indices para obtener una vista como la esperada
tabla_resumen.set_index(['race/ethnicity','simple parental level of education'],inplace=True)
tabla_resumen
````

````python
#Cargar los datos other_grades
path3= path+"other_grades.csv"
df_grades3 = pd.read_csv(path3)
df_grades3
````

````python
#Outer join
pd.merge(
    left=df_grades,
    right=df_grades3,
    on="names",
    how="outer",
    sort=True,
)
````

Al utilizar outer join lo que se hace es combinar ambos dataframe, manteniendo tanto los nombres que tienen en común como los que no. Es por ello que en la cantidad total de filas se siguen teniendo 1000, ya que habían 875 en común y 125 extra que se mantuvieron del data ```other_grades```, estas 125 presentan varios valores NaN ya que no se pueden completar las columnas que no estaban presentes en su dataframe original.

````python
#Left join
pd.merge(
    left=df_grades,
    right=df_grades3,
    on="names",
    how="left",
    sort=True,
)
````

En el caso del left join se conservan sólo los elementos que han sido combinados exitosamente del dataset left, que en este caso es ```df_grades```. Por lo que se obtienen 875 datos que son todos los del dataset ```df_grades``` y que se pudieron combinar correctamente con alguno en el dataset nuevo, así, no quedan valores NaN en el join.

````python
#Right join
pd.merge(
    left=df_grades,
    right=df_grades3,
    on="names",
    how="right",
    sort=True,
)
````

En el caso del right join se conservan sólo los elementos que están en el dataset right, que en este caso es ```df_grades3``` que contienen las nuevas notas. Dado que ```df_grades3``` contiene los datos duplicados, se obtienen 1000 datos, de los cuales sólo 875 se combinaron correctamente, mientras que 125 datos poseen NaN en sus columnas, ya que no se pudieron combinar exitosamente.

````python
#Inner join
pd.merge(
    left=df_grades,
    right=df_grades3,
    left_on="names",
    right_on="names",
    how="inner",
    sort=True,
)
````

Al hacer inner join se combinan los datos que se encuentran en ambos dataframes, descartando todo el resto. Con lo cual se obtiene un dataset con los 875 datos deseados y se han descartado los 125 extra que poseía el nuevo dataframe. Es el mismo resultado que se obtuvo al realizar left join.

> **Justificación:**


En este caso, ya que un dataframe está contenido en el otro (al menos en la columna names que es la utilizada para realizar el join), hacer left join o inner join resulta equivalente, por lo que cualquiera de estas resultaría para el propósito que se tiene.  

En otro caso, que no hubiese un dataset contenido en otro, habría que decidir cuál de ambos efectos es el que se está buscando. Por ejemplo si se supiera que ```df_grades``` tiene los nombres de todos los estudiantes, pero no se encontrara contenido en ```df_grades3``` al hacer inner join se perderían datos de algunos estudiantes (los que no están contenidos), por lo que sería más conveniente realizar el left join; podrían haber más casos posibles que no daría lo mismo y se deberían estudiar con detención.
