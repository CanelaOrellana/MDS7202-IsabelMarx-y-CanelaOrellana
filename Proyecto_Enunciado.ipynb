### 2.3 Unir dataframes

A continuación se unen los dataset ```df_1``` y ```df_2```, dado que no necesariamente son disjuntos, luego de unirlos se deben eliminar los duplicados, priorizando la permanencia de aquellos que vienen de ```df_1```ya que se sabe que es más confiable.  

````python

df_3 = pd.concat([df_1,df_2], axis=0)
df_3.head(5)
````

````python
df_3.drop_duplicates(keep='first', inplace=True)#, ignore_index=False)
df_3
````

El dataset ```df_3``` corresponde a la unión entre ```df_1``` y ```df_2```, en vista de que este contiene 934.858 filas, se concluye que existían 202.334 filas duplicadas entre ambos datasets.

### 2.4 Unir nuevas variables

En la siguiente celda, se combinan los dataset ```df_3``` y ```df_email```, en vista de que ambos datasets no contienen las mismas columnas, se debe realizar un _merge_ entre ambos, de modo que se combine la información que contienen. Para que permanezcan los datos de ambos dataset se utiliza el método _outer_.

````python
df = pd.merge(
    left=df_3,
    right=df_email,
    on="id",
    sort=True,
    how="outer",
)
df.head(5)
````

Se puede observar que en el ```id``` 1 se presentó un dato que no estaba presente en los ejemplos originales, pero sí en las características de email y teléfono contenidas en el otro dataset.

### 6.4 Calcular ratios y responder

````python
df_jov_sample['predict'] = outl_jov
df_ajov_sample['predict'] = outl_ajov
df_adul_sample['predict'] = outl_adul
df_pmay_sample['predict'] = outl_pmay
````

````python
#df_jov_sample[df_jov_sample['1']==0.0]
df_jov_sample['1'].sum()
````

````python
#Datos necesarios
Fraude_pred_bien_j = len(df_jov_sample[(df_jov_sample['1']==1) & (df_jov_sample['predict']==-1)])
Fraude_pred_j = len(df_jov_sample[df_jov_sample['predict']==-1])
Fraude_real_j = len(df_jov_sample[df_jov_sample['1']==1])

Fraude_pred_bien_aj = len(df_ajov_sample[(df_ajov_sample['1']==1) & (df_ajov_sample['predict']==-1)])
Fraude_pred_aj = len(df_ajov_sample[df_ajov_sample['predict']==-1])
Fraude_real_aj = len(df_ajov_sample[df_ajov_sample['1']==1])

Fraude_pred_bien_a = len(df_adul_sample[(df_adul_sample['1']==1) & (df_adul_sample['predict']==-1)])
Fraude_pred_a = len(df_adul_sample[df_adul_sample['predict']==-1])
Fraude_real_a = len(df_adul_sample[df_adul_sample['1']==1])

Fraude_pred_bien_pm = len(df_pmay_sample[(df_pmay_sample['1']==1) & (df_pmay_sample['predict']==-1)])
Fraude_pred_pm = len(df_pmay_sample[df_pmay_sample['predict']==-1])
Fraude_real_pm = len(df_pmay_sample[df_pmay_sample['1']==1])

#Cálculo de P
P_j = Fraude_pred_bien_j/Fraude_pred_j
P_aj = Fraude_pred_bien_aj/Fraude_pred_aj
P_a = Fraude_pred_bien_a/Fraude_pred_a
P_pm = Fraude_pred_bien_pm/Fraude_pred_pm

#Cálculo de R
R_j = Fraude_pred_bien_j/Fraude_real_j
R_aj = Fraude_pred_bien_aj/Fraude_real_aj
R_a = Fraude_pred_bien_a/Fraude_real_a
R_pm = Fraude_pred_bien_pm/Fraude_real_pm

print('El valor obtenido para P en los jóvenes es: '+ str(P_j) + '\n El valor obtenido para R en los jóvenes es: '+ str(R_j))
print('El valor obtenido para P en los adultos jóvenes es: '+ str(P_aj) + '\n El valor obtenido para R en los adultos jóvenes es: '+ str(R_aj))
print('El valor obtenido para P en los adultos es: '+ str(P_a) + '\n El valor obtenido para R en los adultos es: '+ str(R_a))
print('El valor obtenido para P en las personas mayores es: '+ str(P_pm) + '\n El valor obtenido para R en las personas mayores es: '+ str(R_pm))
````

Los valores P y R que se solicitan corresponden a las métricas _'Precision'_ y _'Recall'_ respectivamente. Estos se interpretan según:

- Precision: se encuentra entre 0 y 1. Se puede pensar como el porcentaje de los datos predichos como fraude, que realmente eran fraude. Por ello, mientras más cercano a 0, la predicción realizada es peor, ya que pocos de los datos predichos como fraude son realmente fraude. Por otra parte, mientras más cercano a 1 se tiene que la mayoría de los predichos como fraude, realmente lo son, lo cual es lo que se busca al predecir.

- Recall: también se encuentra entre 0 y 1. Del mismo modo que en _Precision_ , mientras más cerca de 0, la predicción es peor. La diferencia es que en este caso, se interpreta como el porcentaje de los fraudes totales, que se logró predecir.

Sabiendo lo anterior, los valores obtenidos para Precision y Recall al identificar los outliers (anomalías) como fraudes, son predicciones bastante malas. Esto se puede deber a que si bien se espera que los fraudes no sean comunes, no se debe asumir que aparecerán como outliers en los datos, primero que todo porque por lo general, los outliers son producidos por errores en la medición de los datos, por lo que más que detectar fraudes, se detectan outliers clásicos de los datos. Por ello, es coherente haber obtenido malos resultados al utilizar esta estrategia de predicción. 

Los mejores resultados se obtuvieron en los segmentos etarios de 'Adultos jóvenes' y de 'Adultos'. Esto podría deberse a que son aquellos cuyas muestras quedaron más grandes, pero por otro lado, cabe mencionar que estos resultados cambian cada vez que se vuelve a correr el modelo y se obtiene una muestra diferente, por lo que es posible que al momento de escribir este párrafo se haya obtenido una 'muestra suertuda' de esos dataframes.
